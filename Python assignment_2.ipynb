@.1 Discuss string slicing and provide examples.

Ans - String slicing is like cutting out a piece of that line. We can choose exactly where to start the cut and where to end it.

For example = string[start:end:step]

Start and End: We pick two points on the line. The letter at the first point is included in the cut, but the letter at the second point is not.

Step Size: We can also decide how many letters to skip between each cut. For example, you could take every other letter or every third letter.

```python
#Example 1.

name = "kapil kumar"
substring = name[0:5]  # Extract characters from name 0 to 4.
print(substring)

##Example 2.

phone_no = "9818336336"  ## Extract odd place digits from phone_no.
odd_digits = phone_no[::2]
print(odd_digits)
```

@.2 Explain the key features of lists in Python.

Ans - A list is like a digital shopping list. You can write down different kinds of things in it:- numbers, words, or even other lists.

Here are the key features:- Ordered, Mutable, Heterogeneous, Indexing and Slicing.

Here's what we can do with a Python list:-

1. Adds an item to the end of the list. -->> append():
2. Inserts an item at a specific index. -->> insert():
3. Removes the first occurrence of an item. -->> remove():
4. Removes and returns the item at a specific index. -->> pop():
5. Sorts the list in ascending order. -->> sort():
6. Reverses the order of the items. -->> reverse():

```python
#Example

my_list = [1, 7, "kapil", "apple", "goa", 3.14, "banana"]

# Accessing elements:
print(my_list[0])
print(my_list[2])

# Adding an element:
my_list.append("kumar")
print("after add = ", my_list)

# Removing an element:

my_list.remove(3.14)
print("after remove = ", my_list)

# Sorting the list:

my_new_list = [1, 7, 45, 5555, 254, 4, 5, 98, 65, 25]
my_new_list.sort()
print(my_new_list)
```

@.3 Describe how to access, modify, and delete elements in a list with examples.

Ans - Access elements:- To access a specific element in a list, we use its index. The index starts from 0 for the first element.

Modify elements:- To modify an element, we assign a new value to it using its index.

Delete Elements:- we can delete elements using the del keyword or the pop() method.



```python
#Example

list = [5, 66, 85, 98, 10, 65, 74, 999, 65, 36, 336,]

#---------Access elements-----------

print(list[0])  # Accessing the first element

print(list[2])  # Accessing the third element

print(list[-1]) # Accessing the last element

#---------Modify elements-----------

list1 = [5, 66, 85, 98, 10, 65, 74, 999, 65, 36, 336,]

# Changing the second element to 25:

list1[1] = 25
print("Modify list =", list1)

#---------Delete elements-----------

list2 = [5, 66, 85, 98, 10, 65, 74, 999, 65, 36, 336,]

# Deleting the third element:

del list2[2]
print("delete list = ", list2)
```

Ans - A set is a collection of unique objects.

Key Features of Sets:-

1. Unordered:- Items in a set don't have a specific order.

2. Unique:- A set cannot contain duplicate items.

3. Mutable:- We can add or remove items from a set after it's created.


```python
#Example:-

my_set = {1, 2, 3, 4, 5} #In this set, each number appears only once, and the order of the numbers doesn't matter.

#Uses of Sets:-

#Removing Duplicates

list = [1, 2, 2, 3, 3, 3,4,4,4,5,5,8,9,47]
my_set = set(list)
print(my_set)

#Checking Membership:

list2 = [1, 2, 2, 3, 3, 3,4,4,4,5,5,8,9,47]
my_new_set = set(list2)

if 5 in my_new_set:
    print("5 is in the set")
```

@.8 Discuss the importance of dictionary keys being immutable and provide examples.

Ans - In Python, dictionary keys must be immutable objects. 

Immutable keys ensure consistency. Once a key-value pair is added to a dictionary, the key's identity remains fixed. This prevents unexpected behavior and errors that might arise from changing the key's value. 

This is a crucial requirement for the efficient functioning of dictionaries because dictionaries use a hashing algorithm to store and retrieve values efficiently. The hash value of a key is calculated based on its content. If a key were mutable, its content could change, and thus its hash value could also change.


```python
##Examples::--

#string:-

dict = {"apple": 3, "banana": 2, "orange": 5}
print(dict)
type(dict)

#Integers:-

my_marks = {100: 95, 50: 40, 200: 192}
print(my_marks)
type(my_marks)

#Tuples:-

my_info = {(1990, 15, 8): "kapil", (1992, 5, 15): "kumar"}
print(my_info)
type(my_info)
```

```python
##Examples::-- Mutable Objects Not Suitable for Dictionary Keys

#Lists

my_dict1 = {[1, 2, 3]: "list_key"}   # This will raise a TypeError
print(my_dict1)

#Dictionaries

my_dict2 = {{1: "a"}: "dict_key"} # This will also raise a TypeError
print(my_dict2)
```
