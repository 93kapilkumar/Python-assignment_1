@.1 What is the difference between a function and a method in Python?

Ans - A function and a method in Python are both used to do a specific task, but they have some differences:

Function:- A function is not tied to any object; it works independently. We can use it anywhere in your program by calling its name.

Method:- A method is a function that belongs to an object (usually inside a class). It works with the data of that object.


```python
#Function Example:-

def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(result)
```

```python
#Method Example:-

class Calculator:
    def add(self, a, b):  
        return a + b

calc = Calculator()
result = calc.add(9, 3)
print(result)
```

@.3 What are the different ways to define and call a function in Python?

Ans - In Python, We can define and call functions in different ways based on how they are designed. Below is a simple explanation with examples.

1. Regular Function:- No parameters, just call it.
2. With Parameters:- Pass data when calling.
3. Return Value:- Get output from the function.
4. Default Parameters:- Use default values if no input is provided.
5. Variable-Length Arguments:- Handle many inputs with *args or **kwargs.
6. Lambda Function:- Quick one-line functions.


```python
#1. Regular Function:- A regular function is defined using the def keyword and can be called by its name.

#Example:-

def greet():
    print("Hello!")

greet()
```

@.5 What are iterators in Python and how do they differ from iterables?

Ans - Iterable:- An iterable is any object that you can loop through (like a list, string, or tuple). Examples of iterables: list, tuple, string, dictionary, etc.

To get values from an iterable one at a time, you need an iterator.

Iterator:- An iterator is an object that keeps track of where it is during iteration. We can get the next value from an iterator using the next() function. Once all values are used up, the iterator raises a StopIteration error.

How Are They Different?

Aspect Iterable Iterator Definition:- An object you can loop through. An object used to get values one at a time from an iterable.Example:- list, tuple, string (e.g., [1, 2, 3])	

Created using iter() (e.g., iter([1, 2, 3])).

Access Values:- We Can use a for loop directly.	Use next() to access values one at a time.



```python
#Example 1: Iterable

my_list = [1, 2, 3]

for item in my_list:
    print(item)
```

```python
#Example 2: Iterator

my_list = [1, 2, 3]
my_iterator = iter(my_list)   # Convert the list (iterable) into an iterator

print(next(my_iterator))      # Use the next() function to get values one at a time
print(next(my_iterator))  
print(next(my_iterator))      # If you call next() again, it will raise a StopIteration error.
```

@.6 Explain the concept of generators in Python and how they are defined.

Ans - A generator in Python is a special type of function that produces values one at a time instead of creating them all at once. It is useful when working with large data because it doesn't use up memory to store all the values.

How to Define a Generator:- 

1. A generator is defined like a normal function using the def keyword.        
2. Instead of return, it uses the yield keyword to produce values one by one.
   
Key Features of Generators:- 

Produces Values Lazily:- Generators only produce the next value when asked, saving memory.

Iterable:- Generators can be used in loops or with the next() function to get values.

Stops Automatically:- When all values are generated, it stops with a StopIteration error.

```python
#Example: Simple Generator

def number_generator():
    yield 1  # Produces the first value
    yield 2  # Produces the second value
    yield 3  # Produces the third value

# Use the generator
gen = number_generator()

print(next(gen))  
print(next(gen))  
print(next(gen))  
# If you call next() again, it will raise StopIteration
```

```python
#Example: Using a Generator in a Loop

# Define a generator
def countdown(n):
    while n > 0:
        yield n  # Produce the current value of n
        n -= 1   # Decrease n by 1

# Use the generator in a for loop
for num in countdown(5):
    print(num)
```

@.7 What are the advantages of using generators over regular functions?

Ans- Generators in Python have several advantages compared to regular functions, especially when handling large data or complex computations. Below is a simple explanation:-

1. Saves Memory:- Generators dont store all data at once.
2. Faster for Large Data:- Produces values on-demand.
3. Simpler Code:- Makes iteration logic cleaner and easier.
4. Handles Infinite or Large Sequences:- Useful when you dont know the total size.
5. Supports Lazy Evaluation:- Processes data only when needed.

```python
#Example:

# Generator example
def even_numbers(limit):
    for i in range(limit):
        if i % 2 == 0:
            yield i

for num in even_numbers(10):
    print(num)
```

@.10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans - map(): Transforms each item in an iterable.

reduce(): Reduces an iterable to a single value.

filter(): Filters items in an iterable, keeping only those that meet a condition.


```python
#Using map() to Square Each Number:

numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))

#Using reduce() to Find the Product of Numbers:

from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  

#Using filter() to Find Even Numbers:

numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens)) 
```

@.1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.


```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

numbers_list = input("Enter the numbers separated by commas: ").split(',')

numbers_list = [int(num.strip()) for num in numbers_list]

result = sum_even_numbers(numbers_list)

print("Sum of even numbers:", result)
```

@.4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True 

for number in range(1, 201):  
    if is_prime(number):
        print(number, "is prime")
```

@.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  
        self.a, self.b = 0, 1  
        self.count = 0 

    def __iter__(self):
        return self  

    def __next__(self):
        if self.count >= self.terms:  
            raise StopIteration
        else:
            fib_number = self.a  
            self.a, self.b = self.b, self.a + self.b  
            self.count += 1  
            return fib_number 

num_terms = 10  
fib_iterator = FibonacciIterator(num_terms)

for number in fib_iterator:
    print(number)
```

@.6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i  

exponent = 5  
for power in powers_of_two(exponent):
    print(power)
```

@.7 Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

file_path = 'Desktop/photoshop.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

@.9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temps = [0, 20, 37, 100, -10]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```
