@.3 Compare and contrast mutable and immutable objects in python and with examples.

Ans. Mutable Objects: --> Can be changed after creation.
                      --> Modifications are made directly to the object's memory location.
                      --> Examples: Lists, dictionaries.
     Immutable Objects: --> Cannot be changed after creation.
                        --> Any operation that appears to modify an immutable object actually creates a new object.
                        --> Examples: Numbers, strings, tuples.

```python
#Examples

#mutabile object

fruit_list = ["apple", "banana", "orange"]
print(id(fruit_list))
print(fruit_list)
print(type(fruit_list))
```

```python
# Adding an item to the list:
fruit_list.append("grape")
print(id(fruit_list))
print(fruit_list)
print(type(fruit_list))
```

```python
# Removing an item from the list:
fruit_list.remove("banana")
print(id(fruit_list))
print(fruit_list)
print(type(fruit_list))
```

```python
#See fruit_list value is same in every condition this is called mutable object.
```

```python
#Immutabile object

a = 10
b = 10
c = 10

print (id(a))
print (id(b))
print (id(c))
print ("a -", a)
print ("b -", b)
print ("c -", c)
print (type(a))
```

See they have same value for variable a,b,c


```python
a = 10
b = 10
c = 10

# Now change the value of variable 'a'

a = 20+a

print (id(a))
print (id(b))
print (id(c))
print ("a -", a)
print ("b -", b)
print ("c -", c)
print (type(a))
```

```python
#Now see variable 'a' value is change but variables b and c values are same this is called immutable objects.
```

@.4 Discuss the different types of operators in python and provide examples of how they are used.

Ans. Think of operators as tools that perform specific actions on numbers or values. 
     There are some common ones:- 1. Arithmetic Operators
                                  2. Comparison Operators
                                  3. Logical Operators
                                  4. Assignment Operators
                                  5. Bitwise Operators

```python
#Arithmetic Operators

x = 10
y = 5

# Addition
print(x + y)

# Subtraction
print(x - y)

# Multiplication
print(x * y)

# Division
print(x / y)

# Modulus (remainder)
print(x % y)

# Exponentiation
print(x ** y)
```

```python
#Comparison Operators

x = 10
y = 5

# Equal to
print(x == y)

# Not equal to
print(x != y)

# Greater than
print(x > y)

# Less than
print(x < y)

# Greater than or equal to
print(x >= y)

# Less than or equal to
print(x <= y)
```

```python
#Logical Operators

x = True
y = False

# AND operator
print(x and y)

# OR operator
print(x or y)

# NOT operator
print(not x)
```

```python
# Addition assignment

x = 10
y = 5

x += y  # x = x + y
print(x)

# Subtraction assignment
x -= y  # x = x - y
print(x)

# Multiplication assignment
x *= y  # x = x * y
print(x)

# Division assignment
x /= y  # x = x / y
print(x)

# Floor division assignment
x //= y  # x = x // y
print(x)

# Modulus assignment
x %= y  # x = x % y
print(x)

# Exponentiation assignment
x **= y  # x = x ** y
print(x)
```

```python
#Bitwise Operators

x = 10
y = 5

print(bin(x))
print(bin(y))

# Bitwise AND
print(x & y)

# Bitwise OR
print(x | y)

# Bitwise XOR
print(x ^ y)

# Bitwise NOT
print(~x)

# Left shift
print(x << 2)

# Right shift
print(x >> 2)
```

@.6 How to conditional statement work in python? Illustrate with examples.

Ans. Conditional statements in Python are used to control the flow of execution based on certain conditions. They allow we to make decisions within our program.

Key Conditional Statements:- 1.if statement 2.if-else statement 3.if-elif-else statement 

```python
#if statement

age = int(input("Write here your age"))

if age>=18:
    print("You are eligible for vote")
```

```python
#.if-else statement

age = int(input("Write here your age"))

if age>=18:
    print("You are eligible for vote")
else:
    print("You are not eligible for vote")
```

```python
#if-elif-else statement

# Check if a number is a single-digit number, 2-digit number, and so on, up to 5 digits:

N = int(input("Here type a number between 0 to 99999: "))

if N >=0 and N<=9:
    print("this is 1 digit no")
elif N >=10 and N<=99:
    print("this is 2 digit no")
elif N >=100 and N<=999:
    print("this is 3 digit no")
elif N >=1000 and N<=9999:
    print("this is 4 digit no")
elif N >=10000 and N<=99999:
    print("this is 5 digit no")
else:
    print("you type more than 5 digit number")
```

@.7 Describe the diffrent types of loops in python and their use cases with examples.

Ans. Python provides two main types of loops: for loops and while loops. 
These loops are used to execute a block of code repeatedly based on certain conditions.

```python
#for Loop:- Used to iterate over a sequence of items, such as a list, tuple, string, or range of numbers.

number = 5
factorial = 1

for i in range(1, number + 1):
    factorial *= i

print("The factorial of", number, "is", factorial)
```

```python
#while Loop:- Executes a block of code repeatedly as long as a given condition is True.

coins = 0

while coins < 10:
  print("Please Earn more coins.")
  coins += 1

print("You have enough coins! Enjoy your money!")
```
